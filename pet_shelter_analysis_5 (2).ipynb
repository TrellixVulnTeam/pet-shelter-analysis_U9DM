**5)What are the adoption rates for specific breeds?
Find the top 5 dog breeds in the shelter (based on count) and then find the adoption percentage of each breed.**

```python
breeds_series = intakes.loc[intakes['animal_type']=='Dog'].groupby('breed').size().sort_values(ascending=False)[:5]
```

```python
breeds_frame = pd.DataFrame(breeds_series)
```

```python
breeds_frame = breeds_frame.rename(columns={0: 'Count'})
```

```python
# Calculate Percentage
breeds_frame['adoption_percent'] = (breeds_frame['Count'] / 
                      adopted_pets['breed'].count()) * 100
# Show the dataframe
breeds_frame
```

**The table above shows the adoption percentage of each breed. To find this result I filtered out the dogs in the intakes database, grouped the results by breed and sorted the data by number of intakes per breed. Then I put this series into a dataframe so that I could better manipulate the data. After doing this I divided the number of intakes per breed by the total number of adopted pets in the shelter and multiplied the result by 100**


**6) What are the adoption rates for different colorings?**
Find the top 5 colorings in the shelter (based on count) and then find the adoption percentage of each color.

```python
colors_series = intakes.groupby('color').size().sort_values(ascending=False)[:5]
colors_frame = pd.DataFrame(colors_series)
colors_frame = colors_frame.rename(columns={0: 'Count'})
# Calculate Percentage
colors_frame['adoption_percent'] = (colors_frame['Count'] / adopted_pets['color'].count()) * 100
# Show the dataframe
colors_frame
```

**The table above contains the answer to the question. I found this in a similar way to how I found the answer to the previous question except I did not include a filter to only analyze the dogs in the database.**

**7)About how many animals are spayed/neutered each month?**
This will help the shelter allocate resources and staff. Assume that all intact males and females will be spayed/neutered.

```python
# Filters intakes by the intact females and males
intact = intakes.loc[((intakes['sex_upon_intake']=='Intact Female') | (intakes['sex_upon_intake']=='Intact Male' ))]
```

```python
# N.B that 'table' is a DataFrame
table = pd.pivot_table(intact, values='animal_id', index=['month_found'],
                    columns=['year_found'], aggfunc='count')
#Finds the average number of intakes per month over the years
table['Average'] = table.mean(axis = 1)
# Rename the months with month names instead of numbers
table = table.rename(index={1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June'})
table = table.rename(index={7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})
table.round()
```

**The table above contains the average number of animals that were spayed/neutered per month in the right-most column. I found this result by first creating a pivot table to sort the number of animals brought into the shelter for each month and year(I had already added 2 additional columns to the intake data set with the year and the month that each animal was found). After, I added another column titled "Average" that calculated the average across each row to show the average number of animals that were brought into the shelter for a particular month over the years.**
